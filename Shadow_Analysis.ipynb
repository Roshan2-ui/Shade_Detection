```python
import os
import arcpy
from arcpy import env
from datetime import datetime
```

```python
gdb_path = r"D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb"
arcpy.env.workspace = gdb_path
arcpy.env.overwriteOutput = True
arcpy.ImportToolbox(r"D:\My_project\MXD\ShadowImpactAnalysis\ShadowImpactAnalysis.tbx", "ShadowTool")
```

```python
S_Point= datetime.now()
Sun_Map = "Sun_Position_5"
AOI_fc = r"D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb\Final_Study_Area_Poly"
study_area_3d = r"D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb\Study_Area_Final_v2"

F_Tree = "Final_Tree"
F_Building = "Final_Building_v7_1"

arcpy.MakeFeatureLayer_management(F_Building, F_Building)
arcpy.MakeFeatureLayer_management(F_Tree, F_Tree)
arcpy.MakeFeatureLayer_management(study_area_3d, study_area_3d)
```

```python
#Shade_Analysis_2.0

for i in range(1,174):
    
    TSTART = datetime.now()
    
    arcpy.conversion.FeatureClassToFeatureClass(AOI_fc,gdb_path,"UID_lyr","UID = {}".format(i))
    arcpy.conversion.FeatureClassToFeatureClass(study_area_3d,gdb_path,"lyr_3d","UID = {}".format(i))
    
    temp_AOI = r"D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb\temp_AOI"
    
    arcpy.management.SelectLayerByLocation(F_Building, "HAVE_THEIR_CENTER_IN","UID_lyr"," ", "NEW_SELECTION")
    print("Selection Done !!")
    
    arcpy.CopyFeatures_management(F_Building, "D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb\Building_M")
    
    print("Layer Creation Done !!")
    
    arcpy.management.Merge(["Building_M","lyr_3d"], "temp_model")
    print("Temporary Model done!!")
    
    arcpy.analysis.Buffer("UID_lyr","Buffed_AOI", "150 Meters")
    print("Building_S and Tree_S Creation Started...")
       
    arcpy.management.SelectLayerByLocation([F_Building, F_Tree], "HAVE_THEIR_CENTER_IN","Buffed_AOI"," ", "NEW_SELECTION")
    arcpy.CopyFeatures_management(F_Building, "D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb\Building_S")
    arcpy.CopyFeatures_management(F_Tree, "D:\My_project\MXD\Trial_for_coding\Datatset\Shadow_Analysis_2.gdb\Trees_S")
    print("All pre requirements Done !!")
    
    temp_model = "temp_model"
    Tree = "Trees_S"
    Building = "Building_S"
    Shade_1_ = r"D:\\My_project\\MXD\\Trial_for_coding\\Datatset\\Shadow_Analysis_2.gdb\\Shade_Combined_{}_5m".format(i)
    arcpy.ShadowTool.ShadowAnalysis (
        Input_Features=temp_model, 
        Output_Features = Shade_1_, 
        Panel_Size = 5, 
        Sun_Features = Sun_Map,
        #Obstruction_Features = Building,
        #Obstruction_Features = Tree,
        Obstruction_Features = [Tree, Building],
        Use_Observer_Features_as_Obstruction = False
        )
    TEND = datetime.now()
    print(f'Processing time: {TEND-TSTART} [HH:MM:SS]')
    print("Processed Model Number {} ".format(i))
        
    print(" ")
    
E_Point= datetime.now()
print(f'Processing time: {E_Point-S_Point} [HH:MM:SS]')
```
